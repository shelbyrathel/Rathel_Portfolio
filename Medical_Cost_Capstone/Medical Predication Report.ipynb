# <ins> Predictions for Setting Insurance Premiums </ins>

Medical costs in the U.S. have been extremely high and have been rising for decades. In 2023, the country spent 4.87 trillion on healthcare, which amounts to approximately 14,570 per person. This figure represents a 7.5% increase from the previous year.

## <ins> 1. Data </ins>

For this project, an online dataset from [Kaggle](https://www.kaggle.com/datasets/mirichoi0218/insurance/data) was utilized, which includes variables such as age, sex, body mass index, the number of dependents covered by insurance, smoking status, region of residence, and costs billed to insurance. The dataset is designed for instructional purposes, allowing the assumption that all individuals come from the same country and have a consistent billing date.

The stakeholders in this project include strategic decision-makers within an insurance company. They will use the predictions to set premiums, manage risks, and ensure profitability while remaining competitive. The analysis will focus on identifying trends in the data to predict insurance billing costs based on the provided variables.

The link to the complete data notebook is [here](https://github.com/shelbyrathel/Rathel_Portfolio/blob/main/Medical_Cost_Capstone/Wrangling%2CPreprocessing%2C%26Modeling.ipynb). 

## <ins> 2. Data Cleaning </ins>

The data was initially downloaded from Kaggle, then read into the notebook. By looking at the data values. There were no missing values, and all data was of the correct type. There were no duplicates of any rows. 

## <ins> 3. EDA </ins>

The data shows that there were more values recorded in the southeastern region than in any other area. This trend may be attributed to a higher population in that region; however, without specific data, this cannot be definitively confirmed. The number of male and female individuals was approximately equal. Additionally, there were significantly more nonsmokers than smokers, with a count of 1,063 nonsmokers compared to 274 smokers. This aligns with the fact that only 11.6% of individuals in the United States were smokers in 2022.
Upon examining the distribution of various characteristics, it is evident that all values are clustered within similar ranges. The outliers in insurance costs do not indicate inappropriate values; rather, they represent extremely high costs. Notably, there are several outliers among individuals with a higher number of children when compared to those with fewer children. Interestingly, individuals with no children incur the highest charges. This observation aligns with the general belief that families with more children are less likely to undergo expensive procedures, which ultimately translates to lower costs for insurance companies. In terms of geographical regions, there appears to be no significant difference in the distribution of costs. However, there is a marked difference in expenses between smokers and non-smokers, with smokers generally facing higher costs. It is worth noting that non-smokers are the only group that exhibits outliers in this analysis, yet smokers still have the highest overall costs. Furthermore, there is no discernible difference in costs between genders.

## <ins> 4. Algorithms & Modeling Building</ins>

The analysis of the data reveals that charges are a continuous variable. To model the dataset effectively, the characteristics were converted into binomial indicators, allowing for the use of regression techniques. The dataset was then divided into training and testing sets, with 85% of the data allocated for training and 15% for testing. Several potential regression models were fitted to the data, including Linear Regression, Random Forest, Lasso, K-Nearest Neighbors, Ridge, and Gradient Boosting. The results of these models can be visualized below.

## <ins> 5. Which Model to use? </ins>

To determine the best model, we calculated the Mean Absolute Error, Mean Squared Error, and R values, which were organized into a table for comparison. The table was sorted by the Mean Squared Error.
<img src="Comparisons.png" width="650"/>
From the analysis of the table, it is clear that the best model is Gradient Boosting Regression. This model achieved the lowest Mean Absolute Error, the lowest Mean Squared Error, and the highest R value.

## <ins> 6. Predictions </ins>

Based on this data, we can conclude that physical characteristics do indeed predict health insurance costs. Further analysis reveals that being a smoker, age, and BMI have the most significant impact on insurance costs. Higher values for these factorssuch as being a smoker, being older, or having a high BMIconsistently lead to increased predicted insurance costs. Conversely, individuals who are younger, non-smokers, or have a lower BMI are more likely to have reduced cost predictions. Other variables, such as the number of children, region, and gender, have relatively minimal impacts on the predicted costs according to the model.
